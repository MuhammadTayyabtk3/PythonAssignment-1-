```python
# Assignment 1

import sys
import datetime
import math

print("Twinkle, twinkle, little star,\n\t\tHow I wonder what you are!\n\t\t\t\tUp above the world so high, Like a diamond in the sky\n\t\t\t\tLike a diamond in the sky\n\t\tHow I wonder what you are")

print(sys.version)

now = datetime.datetime.now()
print("Current date and time:",now)

radius = float(input("Enter the radius of the circle: "))
area = math.pi * (radius ** 2)
print(f"The area of the circle with radius {radius} is: {area:.2f}")

first_name = input("Enter your Firstname : ")
last_name = input("Enter your Lastname : ")
print (last_name + " " + first_name)

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
result = num1 + num2
print(result)

sub1 = int(input("Enter marks of Subject 1: "))
sub2 = int(input("Enter marks of Subject 2: "))
sub3 = int(input("Enter marks of Subject 3: "))
sub4 = int(input("Enter marks of Subject 4: "))
sub5 = int(input("Enter marks of Subject 5: "))
total = sub1 + sub2 + sub3 + sub4 + sub5
percentage = total / 500 * 100
if percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "Fail"
print("----- Mark Sheet -----")
print("Total Marks:", total)
print("Percentage:", percentage)
print("Grade:", grade)

num = int(input("Enter Your Number"))
if num % 2 == 0:
      print("Your Number is Odd:", num)
else:
     print("Your Number is Even:", num)
```
